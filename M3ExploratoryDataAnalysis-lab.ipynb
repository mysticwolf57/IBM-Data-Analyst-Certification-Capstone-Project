## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(x=df['Age'])
plt.xlabel('Converted Compensation')
plt.title('Box Plot of Converted Compensation')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1

print('iqr value is: ',iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

print('lower bounds: ' + str(lower_limit))
print('upper bounds: ' + str(upper_limit))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
df[(df['ConvertedComp'] < lower_limit) | (df['ConvertedComp'] > upper_limit)].ConvertedComp.count()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
no_outliers = (df['ConvertedComp'] < (q1 - 1.5 * iqr)) | (df['ConvertedComp'] > (q3 + 1.5 * iqr))
df[no_outliers] = np.nan
df['ConvertedComp'].mean()
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.
